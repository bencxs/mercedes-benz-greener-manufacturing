```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
%matplotlib inline
```

## Loading and Preprocessing Data

```python
# Load training dataset
df_train = pd.read_csv('train.csv')
print "Train dataset dims: ", df_train.shape
# Load test dataset
df_test = pd.read_csv('test.csv')
print "Test dataset dims: ", df_test.shape

display(df_train.head())
display(df_train.describe())

display(df_test.head())
display(df_test.describe())
```

### Create Dummy Variables

```python
#  Drop the ID column for train
df_train = df_train.drop('ID', axis=1)

# Concat train and test set to create dummy variables
df_merged = pd.concat((df_train, df_test))

# Convert categorical variables to dummy variables
df_merged_dummies = pd.get_dummies(
    df_merged, columns=['X0', 'X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X8'])

print df_merged.shape
print df_merged_dummies.shape

# Split back into train and test sets
# Each has 4209 rows
df_train_dummies = df_merged_dummies.iloc[:4209, :]
df_test_dummies = df_merged_dummies.iloc[4209:, :]

# Keep the ID Column separately for test
test_ids = df_test_dummies['ID']

# Drop the ID column for test
df_test_dummies = df_test_dummies.drop('ID', axis=1)

# Drop the y for test
df_test_dummies = df_test_dummies.drop('y', axis=1)

#  Drop the ID column for train, again
df_train_dummies = df_train_dummies.drop('ID', axis=1)

print df_train_dummies.shape
print df_test_dummies.shape
```

 ### Set Features and Labels

```python
features = df_train_dummies.drop('y', axis=1)
labels = df_train_dummies['y'].map(lambda x: np.log(x)) # Convert to natural log to reduce variance
```

## Feature Selection

```python
# Feature Selection
from sklearn.feature_selection import SelectKBest, f_regression

selector = SelectKBest(f_regression, k=100).fit(features, labels)
x_new = selector.transform(features) # not needed to get the score
scores = selector.scores_

sns.set_style('white')
plt.figure(figsize=(18, 3))
plt.bar((range(features.shape[1])), scores)
plt.xticks(range(features.shape[1]), list(features), rotation='vertical')
plt.show()
```

```python
table = zip(list(features), scores)
table.sort(key = lambda t: t[1], reverse=True)

print table[:121]
```

```python
# Clustering to add features
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

cls = KMeans(n_clusters=4, random_state=42, n_jobs=-1)
x_add = cls.fit_transform(x_new)

score = silhouette_score(x_new, cls.labels_, random_state=42)

print x_add.shape
print x_add[:2]
print score
```

```python
x_new = np.hstack((x_new, x_add))

print x_new.shape
print x_new[:2]
```

## Dimensionality Reduction

```python
# Dimensionality reduction
from sklearn.decomposition import PCA

pca = PCA(n_components=60, random_state=42)
pca.fit(x_new)

print "Sum of all principal components: ", pca.explained_variance_ratio_.sum()
print "Principal Components:"
print pca.explained_variance_ratio_
```

```python
x_new = pca.transform(x_new)

print x_new.shape
```

## Train & Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_new, labels, test_size=0.1, random_state=42)

print X_train.shape
print X_test.shape
print y_train.shape
print y_test.shape
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge, BayesianRidge, SGDRegressor, ElasticNet, OrthogonalMatchingPursuit
from sklearn.svm import LinearSVR, SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor, ExtraTreesRegressor
from sklearn.metrics import r2_score, mean_squared_error

def fit_and_predict(reg, X_train, y_train, X_test, y_test):
    model = reg.fit(X_train, y_train)
    pred = reg.predict(X_test)

    print type(reg).__name__
    print "Training R2 Score: ", reg.score(X_train, y_train)
    print "Test R2 Score: ", r2_score(y_test, pred)
    print "MSE: ", mean_squared_error(y_test, pred)
    print ""
    return model
    
fit_and_predict(Ridge(), X_train, y_train, X_test, y_test)
fit_and_predict(BayesianRidge(), X_train, y_train, X_test, y_test)
fit_and_predict(SGDRegressor(), X_train, y_train, X_test, y_test)
fit_and_predict(Lasso(), X_train, y_train, X_test, y_test)
fit_and_predict(OrthogonalMatchingPursuit(), X_train, y_train, X_test, y_test)
fit_and_predict(SVR(), X_train, y_train, X_test, y_test)
fit_and_predict(RandomForestRegressor(), X_train, y_train, X_test, y_test)
fit_and_predict(GradientBoostingRegressor(), X_train, y_train, X_test, y_test)
fit_and_predict(ExtraTreesRegressor(), X_train, y_train, X_test, y_test)
```

```python
# Gridsearch
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
import scipy.stats as st

# BayesianRidge
alpha_1s = st.uniform(1e-6, 1000)
alpha_2s = st.uniform(1e-6, 1000)
lambda_1s = st.uniform(1e-6, 1000)
lambda_2s = st.uniform(1e-6, 1000)
param_grid_br = {'alpha_1': alpha_1s, 
                 'alpha_2': alpha_2s, 
                 'lambda_1': lambda_1s,
                 'lambda_2': lambda_2s}

# OrthogonalMatchingPursuit
n_nonzero_coefs = [None, round(X_train.shape[1]*0.2, 0), round(X_train.shape[1]*0.3, 0)]
normalize = [True]
param_grid_omp = {'n_nonzero_coefs': n_nonzero_coefs,
                  'normalize': normalize}

# SGDRegressor
loss = ['squared_loss', 'huber', 'epsilon_insensitive', 'squared_epsilon_insensitive']
penalty = [None, 'l1', 'l2', 'elasticnet']
learning_rate = ['constant', 'optimal', 'invscaling']
eta0 = [0.01, 0.001, 0.1]
average = [False, True]
param_grid_sgd = {'loss': loss, 
                 'penalty': penalty, 
                 'learning_rate': learning_rate,
                 'eta0': eta0,
                 'average': average}

# GradientBoostingRegressor
# loss = ['ls', 'lad', 'huber', 'quantile']
n_estimators = [100, 200, 300]
max_depth  = [3, 2, 4]
param_grid_gbr = {
                 'n_estimators': n_estimators, 
                 'max_depth': max_depth
                 }


# SVM
C = st.uniform(0.001, 1000)
epsilon  = st.uniform(0.1, 10)
param_grid_svm = {
                 'C': C, 
                 'epsilon': epsilon
                 }

def gridsearchcv(X, y, reg, param_grid, n_folds):
    param_grid = param_grid
    grid_search = GridSearchCV(reg, param_grid, scoring='r2', cv=n_folds)
    grid_search.fit(X, y)
    grid_search.best_params_
    return grid_search.best_params_

def randsearchcv(X, y, reg, param_grid, n_folds):
    param_grid = param_grid
    grid_search = RandomizedSearchCV(reg, param_grid, scoring='r2', cv=n_folds)
    grid_search.fit(X, y)
    grid_search.best_params_
    return grid_search.best_params_

print randsearchcv(X_train, y_train, BayesianRidge(), param_grid_br, 3)
print gridsearchcv(X_train, y_train, GradientBoostingRegressor(random_state=42), param_grid_gbr, 3)
# print gridsearchcv(X_train, y_train, OrthogonalMatchingPursuit(), param_grid_omp, 3)
# print gridsearchcv(X_train, y_train, SGDRegressor(), param_grid_sgd, 3)
print randsearchcv(X_train, y_train, SVR(), param_grid_svm, 3)
```

```python
# Train tuned models
model_br = fit_and_predict(BayesianRidge(
    alpha_2=682.94516471621625, 
    lambda_1=931.72493828040717, 
    lambda_2=188.78948810027777, 
    alpha_1=436.23771997656905), 
                X_train, y_train, X_test, y_test)

'''model_omp = fit_and_predict(OrthogonalMatchingPursuit(
    n_nonzero_coefs=60.0, 
    normalize=True), 
                X_train, y_train, X_test, y_test)'''

model_gbr = fit_and_predict(GradientBoostingRegressor(
    n_estimators=100, 
    max_depth=2, 
    random_state=42), 
                X_train, y_train, X_test, y_test)

'''model_sgd = fit_and_predict(SGDRegressor(
    loss='squared_epsilon_insensitive', 
    eta0=0.01), 
                X_train, y_train, X_test, y_test)'''

model_svm = fit_and_predict(SVR(
    epsilon=0.01,
    C=1), 
                X_train, y_train, X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score, cross_val_predict

# Perform 6-fold cross validation
def cross_val(model, features, labels, scoring, cv):
    scores = cross_val_score(model, features, labels, scoring=scoring, cv=cv)
    print type(model).__name__
    print "Cross-validated scores:", scores
    print "Mean CV Scores:", abs(scores.mean())
    print ""
    
    return abs(scores.mean())
    
score_br = cross_val(model_br, features, labels, scoring='r2', cv=6)
# score_omp = cross_val(model_omp, features, labels, scoring='r2', cv=6)
score_gbr = cross_val(model_gbr, features, labels, scoring='r2', cv=6)
# score_sgd = cross_val(model_sgd, features, labels, scoring='r2', cv=6)
score_svm = cross_val(model_svm, features, labels, scoring='r2', cv=6)
```

```python
# Make cross validated predictions
plt.figure(figsize=(16,4))
cv_pred_br = cross_val_predict(model_br, features, labels, cv=6)
cv_pred_svm = cross_val_predict(model_svm, features, labels, cv=6)
cv_pred_gbr = cross_val_predict(model_gbr, features, labels, cv=6)

ax1 = plt.subplot(1, 3, 1)
plt.scatter(labels, cv_pred_br)
plt.title('BayesianRidge')
plt.xlabel('True Values')
plt.ylabel('Predictions')

plt.subplot(1, 3, 2, sharex=ax1, sharey=ax1)
plt.scatter(labels, cv_pred_svm)
plt.title('SVM')
plt.xlabel('True Values')
plt.ylabel('Predictions')

plt.subplot(1, 3, 3, sharex=ax1, sharey=ax1)
plt.scatter(labels, cv_pred_gbr)
plt.title('GradientBoostingRegressor')
plt.xlabel('True Values')
plt.ylabel('Predictions')
```

```python
r2 = r2_score(labels, cv_pred_br)
mse = mean_squared_error(labels, cv_pred_br)
print "Cross-Predicted MSE:", mse
print "Cross-Predicted R2:", r2
print ""

r2 = r2_score(labels, cv_pred_svm)
mse = mean_squared_error(labels, cv_pred_svm)
print "Cross-Predicted MSE:", mse
print "Cross-Predicted R2:", r2
print ""

r2 = r2_score(labels, cv_pred_gbr)
mse = mean_squared_error(labels, cv_pred_gbr)
print "Cross-Predicted MSE:", mse
print "Cross-Predicted R2:", r2
print ""
```

```python
# Train final model
from time import time

def build_and_train_ensemble(reg, reg_final, train_features, labels):
    print "Building ensemble of models..."
    t0 = time()
    df = pd.DataFrame()
    ensemble = []
    
    # Prepare model of ensemble
    for rg in reg:
        model = rg.fit(train_features, labels)
        # Store ensemble models for prediction
        ensemble.append(model)
        pred = rg.predict(train_features)
        df[type(rg).__name__] = pred
    df['y_true'] = labels
    
    # Train final model
    df_train = df.ix[:, :-1]
    df_labels = df['y_true']
    final_model = reg_final.fit(df_train, df_labels)
    tt = time() - t0
    print "Trained in {} seconds.".format(round(tt, 3))
    
    # Plot pred
    fig = plt.figure(figsize=(8,8))
    ax = fig.add_subplot(111)
    ax.set_aspect('auto')
    ax = plt.scatter(labels, pred)
    ax = plt.title('Ensemble')
    ax = plt.xlabel('True Values')
    ax = plt.ylabel('Predictions')
    
    return ensemble, final_model, df


reg = [model_br, model_gbr, model_svm]

reg_final = LinearRegression()

ensemble, final_model, df_ensemble = build_and_train_ensemble(reg, reg_final, x_new, labels)
```

```python
# Make predictions

def make_final_pred(ensemble, final_model, test_features):
    t0 = time()
    df = pd.DataFrame()
    
    # First layer prediction of ensemble
    for en in ensemble:
        pred = en.predict(test_features)
        df[type(en).__name__] = pred
    
    # Final model prediction
    final_pred = final_model.predict(df)
    tt = time() - t0
    print "Prediction made in {} seconds.".format(round(tt, 3))
    
    return final_pred

final_pred = make_final_pred(ensemble, final_model, test_features)
```
